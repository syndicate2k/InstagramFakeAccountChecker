```python
# remove this feature from consideration
df = df.drop(columns=['Mutual Friends'])
```

```python
# cuz there's a conflict between 'yes' and 'Yes'
df['Bio'] = df['Bio'].str.lower()
```

```python
strings_columns = ['Bio','Profile Picture', 'External Link', 'Threads']
```

```python
# replace string values ​​one-hot
one_hot_encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = one_hot_encoder.fit_transform(df[strings_columns])
```

```python
df_encoded = pd.DataFrame(encoded, columns=strings_columns)
```

```python
label_encoder = LabelEncoder()
```

```python
# encode the target
df['Labels'] = label_encoder.fit_transform(df['Labels'])
```

```python
df = df.drop(columns=strings_columns)
df = pd.concat([df, df_encoded], axis=1)
```

```python
# somewhere in these columns there were erroneous string values
string_columns = df.select_dtypes(include='object')

print(string_columns)
```

```python
# find a string => replace it with NaN
df['Following/Followers'] = pd.to_numeric(df['Following/Followers'], errors='coerce')
df['Posts/Followers'] = pd.to_numeric(df['Posts/Followers'], errors='coerce')

df.dropna(inplace=True)
```

```python
# for train_test_split and further training, but in my solution it's not used (see further commented lines)
y = df['Labels']
X = df.drop(columns='Labels')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(X.corr(numeric_only=True), annot=True, cmap='Blues')
plt.show()
```

```python
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python
time_limit = 120
```

```python
predictor = TabularPredictor(label='Labels').fit(train_data, presets='best_quality', time_limit=time_limit)
```

```python
y_pred = predictor.predict(test_data.drop(columns=['Labels']))
```
